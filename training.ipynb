```python
import cv2
from ultralytics import YOLO

model = YOLO('runs/detect/train2/weights/best.pt')

cap = cv2.VideoCapture(0)

def process_frame(frame):
    results = model(frame)
    for result in results:
        for detection in result.boxes:
      
            x1, y1, x2, y2 = detection.xyxy[0].tolist()
            
  
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            
       
            label = f'{detection.cls.item():.0f} {detection.conf.item():.2f}'
            font_color = (128, 0, 0)

  
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, font_color, 2)
    
    return frame

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    processed_frame = process_frame(frame)
    cv2.imshow('YOLOv8 Webcam', processed_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('runs/detect/train2/weights/best.pt')

video_file_path = 'IMG_8894.mp4'

output_video_path = 'runs/detect'

cap = cv2.VideoCapture(video_file_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Video codec
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

def process_frame(frame):
    results = model(frame)
    
    for result in results:
        for detection in result.boxes:
            x1, y1, x2, y2 = detection.xyxy[0].tolist()
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            label = f'{detection.cls.item():.0f} {detection.conf.item():.2f}'
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return frame


while True:
    ret, frame = cap.read()
    if not ret:
        break

    processed_frame = process_frame(frame)
    out.write(processed_frame)
    cv2.imshow('YOLOv8 Video Processing', processed_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output video saved to {output_video_path}")

```

```python
import os
import cv2
from ultralytics import YOLO

model = YOLO('runs/detect/train2/weights/best.pt')


test_images_dir = 'test/images'

labeled_images_dir = 'test'
os.makedirs(labeled_images_dir, exist_ok=True) 

test_images = [os.path.join(test_images_dir, filename) for filename in os.listdir(test_images_dir) if filename.endswith('.jpg') or filename.endswith('.png')]

def process_image(image_path, output_dir):
    image = cv2.imread(image_path)
    results = model(image)

    for result in results:
        for detection in result.boxes:

            x1, y1, x2, y2 = detection.xyxy[0].tolist()
        
            cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
      
            label = f'{detection.cls.item():.0f} {detection.conf.item():.2f}'
            
            cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
 
    filename = os.path.basename(image_path)
    output_path = os.path.join(output_dir, filename)
    cv2.imwrite(output_path, image)
    print(f'Saved labeled image: {output_path}')


for image_path in test_images:
    process_image(image_path, labeled_images_dir)

```

```python
import os
import torch
from torchvision.ops import box_iou

def read_bounding_boxes(directory):
    bounding_boxes = {}
    
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            base_name = os.path.splitext(filename)[0]
            
            with open(os.path.join(directory, filename), 'r') as f:
                boxes = []
                
                for line in f:
                    components = line.strip().split()
                    
                    class_id = int(components[0])
                    center_x = float(components[1])
                    center_y = float(components[2])
                    width = float(components[3])
                    height = float(components[4])

                    x_min = center_x - width / 2
                    y_min = center_y - height / 2
                    x_max = center_x + width / 2
                    y_max = center_y + height / 2
                    
                    boxes.append((class_id, x_min, y_min, x_max, y_max))
                
                bounding_boxes[base_name] = boxes
                
    return bounding_boxes


dir1 = 'test/labels'  
dir2 = 'test/coords'  

bounding_boxes1 = read_bounding_boxes(dir1)
bounding_boxes2 = read_bounding_boxes(dir2)


iou_values = []
for file1, boxes1 in bounding_boxes1.items():

    if file1 in bounding_boxes2:

        boxes2 = bounding_boxes2[file1]
        tensor_boxes1 = torch.tensor([box[1:] for box in boxes1], dtype=torch.float32)
        tensor_boxes2 = torch.tensor([box[1:] for box in boxes2], dtype=torch.float32)
        
        for i, box1 in enumerate(boxes1):
            class_id1 = box1[0] 
            for j, box2 in enumerate(boxes2):
                class_id2 = box2[0] 
                
         
                if class_id1 == class_id2:

                    iou = box_iou(tensor_boxes1[i].unsqueeze(0), tensor_boxes2[j].unsqueeze(0))
                    
                    iou_values.append(iou.item())
                    
average_iou = sum(iou_values) / len(iou_values)

print(f"The average IoU: {average_iou:.4f}")
```
